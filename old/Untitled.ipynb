```python
def SingleWindDirection(wind, offset):
    u, v, w = range(3) # for better name of wind variables    
    wind_dir = 180 - (np.arctan2(wind[v], wind[u]) *180 / np.pi)
    if wind_dir >= 360: wind_dir = wind_dir - 360
    if wind_dir < 360: wind_dir = 360 + wind_dir
```

```python
def AngularAverage(ar):
    cossum = (-np.cos(ar / 180 * np.pi)).sum()
    sinsum = (-np.sin(ar / 180 * np.pi)).sum()
    
    cossum = cossum / len(ar)
    sinsum = sinsum / len(ar)
    
    return (np.arctan2(sinsum, cossum) + np.pi) * 180 / np.pi
```

```python
def aa2(ar):
    ar = np.deg2rad(ar)
    x = np.cos(ar).sum()
    y = np.sin(ar).sum()
    return np.rad2deg(np.arctan2(y,x)) % 360
```
